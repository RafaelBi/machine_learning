```python
import matplotlib.pyplot as plt
from numpy.random import exponential
```

```python
# %matplotlib notebook
%matplotlib inline
```

```python
plt.plot(range(20), range(20))
plt.show()
```

```python
plt.scatter(range(20), range(20))
```

```python
sample = exponential(10, 100)
```

```python
plt.plot(sample)
```

```python
plt.hist(sample, bins=20)
plt.show()
```

```python
lineplot = plt.plot(sample,  linewidth=1, color='green')
```

```python
lineplot = plt.plot(sample)
plt.setp(lineplot, linewidth=1, color='green')
```

```python
%matplotlib inline
fig, ax = plt.subplots(1)
lineplot = ax.plot(sample)
ax.set_xlabel('Wow')
ax.set_ylabel('Much')
ax.legend(['First Line'])
ax.set_title('Good old chart', fontdict={'size':30})
ax.set_ylim(-20, 40)
ax.set_xlim(10, 20)
```

```python
import matplotlib.pyplot as plt

sample1 = exponential(100, 100)
sample2 = exponential(10, 1000)

fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=False)

# plt.subplot(2, 1, 1)
axes[0].plot(sample1)
axes[0].set_title('Wow')

# plt.subplot(2, 1, 2)
axes[1].plot(sample2)
plt.show()
```
